```python
import numpy
import matplotlib.pyplot as plt
```

```python
chips = numpy.load('../../chips/64_nw.npy')
```

```python
fig, ax = plt.subplots(figsize=(20,20))
plt.imshow(chips[122])
ax.set_axis_off()
plt.savefig('figs/64.pdf')
```

```python
fig, ax = plt.subplots(figsize=(20,20))
plt.imshow(chips[122][:32, :32, :])
ax.set_axis_off()
plt.savefig('figs/32.pdf')
```
